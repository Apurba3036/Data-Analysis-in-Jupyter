```python
ax=sns.countplot(x='Gender',data=Data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
new_graph=Data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=new_graph)
plt.title('Total Amount Spent by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.show()
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*


```python
ax=sns.countplot(x='Age Group',hue='Gender',data=Data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(x='Age Group',data=Data)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

```python
Sales_State= Data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='State',y='Orders',data=Sales_State)
```

```python
Sales_State= Data.groupby(['State','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x='State',y='Amount',data=Sales_State,hue='Gender')
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


```python
ax=sns.countplot(x='Marital_Status',hue='Gender',data=Data)
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Marital= Data.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
#sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(data = Marital, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

```python
##Occupation
```

```python
ax=sns.countplot(x='Occupation',hue='Gender',data=Data)
sns.set(rc={'figure.figsize':(20,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Job= Data.groupby(['Occupation','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
#sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(data = Job, x = 'Occupation',y= 'Amount', hue='Gender')
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

```python
ax=sns.countplot(x='Product_Category',hue='Gender',data=Data)
sns.set(rc={'figure.figsize':(25,15)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Products= Data.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = Products, x = 'Product_Category',y= 'Amount')
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
Products= Data.groupby(['Product_Category'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = Products, x = 'Product_Category',y= 'Orders')
```

```python
# top 10 most sold products (same thing as above)

fig2, ax2 = plt.subplots(figsize=(10,10))
Data.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='pie')
```

Summary:
Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
